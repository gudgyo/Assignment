```python
url = 'https://raw.githubusercontent.com/qty0/TechTestDigitalyn/main/solar_mesh.csv'
```

```python
df = pd.read_csv(url)
```

```python
df = df.set_index('bins_lat')
```

```python
df.head()
```

We can see that the intervals are relatively uniform, so we dont have to modify the plot.

```python
sns.lineplot(data=[float(x.split()[1][:-1]) - float(x.split()[0][1:-1]) for x in df.index])
```

```python
sns.lineplot(data=[float(x.split()[1][:-1]) - float(x.split()[0][1:-1]) for x in df.columns])
```

The unedited heatmap and the one with the robustified modifier, it yields similar result to the [0-1] normalized one (3rd fig.).

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df, robust=True, ax=ax)
```

```python
normalized_df=(df-df.min())/(df.max()-df.min())
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(normalized_df, robust=False, ax=ax)
```
