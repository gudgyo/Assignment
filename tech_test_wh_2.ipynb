```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/qty0/TechTestDigitalyn/main/solar_mesh.csv'
```

```python
df = pd.read_csv(url)
df = df.set_index('bins_lat')
```

```python
df.head()
```

The task says: "Bin the dataframe's 'weights' into five equeal sized bins, where the bins are counted as = 0 | max/4 | max/2 | max/4*3 | max. Plot this binned data as a histogram using seaborn or plt."

I assume the bin size refers to the intervals of the parameters inside it, otherwise it would be a flat histogram (also it would be impossible to make them equal size with the given intervals).

```python
bins = df.to_numpy().flatten()
```

```python
bins.shape
```

With the given endpoints, there are going to be 4 bins, if we want then to be equal sized.
As we can see, the first interval dwarfs the rest, so it's better to see the second figure which is logarithmic scaled.

```python
sns.set()
h = sns.histplot(bins, bins=[0.0, bins.max()/4, bins.max()/2, (bins.max()/4)*3, bins.max()])
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Bins')

plt.show()
```

```python
sns.set()
h = sns.histplot(bins, bins=[0.0,0.0, bins.max()/4, bins.max()/2, (bins.max()/4)*3, bins.max()], log_scale=[0, 2])
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Bins')

plt.show()
```
